We will perform univariate analysis of <b>funded_amnt, term,int_rate,grade, sub_grade, emp_length, home_ownership, annual_inc, verification_status,loan_status,purpose,addr_state  attributes</b>    

```python
#utility methods
def univariate(dataframe,column,title,bins=15,var_type='cont'):
    if var_type=='cont' :
        plt.figure(figsize=(20,8))
        plt.xlabel(title)
        sns.distplot(pd.Series(dataframe[column],name=title),bins=bins);
        plt.show()
        
    if var_type=='cat' :
        plt.figure(figsize=(20,8))
        plt.xlabel(title)
        dataframe[column].value_counts().plot.bar()
        plt.show()
    
```

# Let's begin with Bilateral analysis

```python
def pivot_table(dataframe,column) :
    pvt = pd.crosstab(dataframe[column], dataframe['loan_status'],margins=True)
    pvt['Prob'] = round((pvt['Charged Off']/pvt['All']),3)
    pvt = pvt[0:-1]
    return pvt
```

```python
def bivariate(dataframe,column,title,stacked=False):
    
    pr_t = pivot_table(dataframe,column)
    display(pr_t)
    ln_Plot = pr_t[['Prob']] 
    line = ln_Plot.plot(figsize=(20,8), marker='o',color = 'g',lw=1)
    line.set_title(title+' vs Default Probablity',fontsize=18,weight="bold")
    #line.set_xlabel(title,fontsize=14)
    line.set_ylabel('Default Probablity',fontsize=20)
    
    barP =  pr_t.iloc[:,0:3] 
    bar = barP.plot(kind='bar',ax = line,rot=1,secondary_y=True,stacked=stacked,alpha=0.3)
    bar.set_ylabel('Number of Applicants',fontsize=20)
    plt.show()
```

```python
bivariate(loan_details,'emp_length','Employee length',True)
```

<b><font color=color>[Insights]</font></b>: Others have higher percentage

```python
bivariate(loan_details,'int_range','Interest Range',True)
```
